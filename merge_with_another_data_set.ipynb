```python
import pandas as pd
```

```python
df_final=pd.read_csv('movies_imdb_final (10K).csv')
```

```python
df_final.columns = [column.strip() for column in df_final.columns]
df_final.columns
```

```python
cols = df_final.select_dtypes(['object']).columns
df_final[cols] = df_final[cols].apply(lambda x: x.str.strip())
```

```python
df_final['Title'] = df_final['Title'].str.upper()
```

```python
df_final.head(2)
```

```python
df=df_final.drop([df_final.columns[0]], axis=1)
```

```python
df.head(1)
```

```python
df_2=pd.read_csv('numbersData.csv')
```

```python
df_2['Title'] = df_2['Title'].str.upper()
```

```python
full_data=df_2.merge(df, on='Title')
```
